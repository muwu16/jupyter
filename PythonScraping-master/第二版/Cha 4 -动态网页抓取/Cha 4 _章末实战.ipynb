```python
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
import random
 
options = Options()
num=str(float(random.randint(500,600)))
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/{}"
                     " (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/{}".format(num,num))
#禁止图片和css加载
prefs = {"profile.managed_default_content_settings.images": 2,'permissions.default.stylesheet':2,"javascript.enabled":False}
options.add_experimental_option("prefs", prefs)
 
driver = webdriver.Chrome(executable_path='/Users/wujiangu/Desktop/chromedriver',options=options)
#把上述地址改成你电脑中geckodriver.exe程序的地址
#在虚拟浏览器中打开 Airbnb 页面
driver.get("https://zh.airbnb.com/s/weihai--China/homes")
time.sleep(2)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

inf_list = []


for i in range(0,15):

    #找到页面中所有的出租房
    rent_list = driver.find_elements_by_css_selector('a._okj6x')
    #对于每一个出租房
    for eachhouse in rent_list:
        #找到评论数量
        try:
            comment = eachhouse.find_element_by_css_selector('span._1clmxfj')
            comment = comment.text
            comment_score = comment.split(" · ")[0]
            comment_num = comment.split(" · ")[1]
        except:
            comment_score = 0
            comment_num = 0

        

        #找到价格
        price = eachhouse.find_element_by_css_selector('div._1ixtnfc')
        price = price.text.replace("每晚", "").replace("价格", "").replace("\n", "")

        #找到名称
        name = eachhouse.find_element_by_css_selector('div._qrfr9x5')
        name = name.text

        #找到房屋类型，大小
        details = eachhouse.find_element_by_css_selector('span._faldii7')
        details = details.text
        house_type = details.split(" · ")[0]
        bed_number = details.split(" · ")[1]
        inf_list.append([comment_score,comment_num, price, name, house_type, bed_number])
#         print (comment_score,comment_num, price, name, house_type, bed_number)
    try:
        next_botton = driver.find_element_by_css_selector('a._za9j7e')
        next_botton.click()
    except:
        break
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```
