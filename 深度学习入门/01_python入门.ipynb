```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
y1 = y
y2 = np.cos(x)
```

```python
plt.plot(x,y1,label = "sin")
plt.plot(x,y2,label = "cos",linestyle = "--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("x&y")
plt.legend()
plt.show()
```
