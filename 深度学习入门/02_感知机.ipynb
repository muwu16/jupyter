```python
def my_and(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    elif tmp >0:
        return 1
```

```python
def my_nand(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) # 仅权重和偏置与AND不同！
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0 
    else:
        return 1
```

```python
def my_or(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) # 仅权重和偏置与AND不同！
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0 
    else:
        return 1
```

## 多层感知机实现异或门

```python
def my_xor(x1, x2):
    s1 = my_nand(x1, x2)
    s2 = my_or(x1, x2)
    y = my_and(s1, s2)
    return y
```
